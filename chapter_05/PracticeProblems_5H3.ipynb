```python
import pymc3 as pm
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import arviz as az
import statsmodels.formula.api as smf
from sklearn import preprocessing
%config InlineBackend.figure_format = 'retina'
az.style.use('arviz-darkgrid')
```

```python
# Load waffle data_labels
df_foxes = pd.read_csv('/home/amit/AmitKushwaha/Courses/Books/StatisticalRethinkingBook/Data/foxes.csv',sep=';')
df_foxes = df_foxes.dropna()
```

```python
df_foxes.head(20).style.bar(subset=['avgfood'], color='lightgreen').bar(subset=['groupsize'], color='red').bar(subset=['area'], color='yellow').bar(subset='weight', color='blue')
```

```python
scaler = preprocessing.StandardScaler()
weight = np.squeeze(scaler.fit_transform(df_foxes['weight'].values.reshape(-1,1)))
area = np.squeeze(scaler.fit_transform(df_foxes['area'].values.reshape(-1,1)))
groupsize = np.squeeze(scaler.fit_transform(df_foxes['groupsize'].values.reshape(-1,1)))
avgfood = np.squeeze(scaler.fit_transform(df_foxes['avgfood'].values.reshape(-1,1)))
```

## Model : weight ~ area + groupsize

```python
with pm.Model() as model_ag:
    a = pm.Normal('a', mu=10, sigma=10)
    ba = pm.Normal('ba', mu=0, sigma=1)
    bg = pm.Normal('bg', mu=0, sigma=1)
    sigma = pm.Uniform('sigma', lower=0, upper=10)
    mu = pm.Deterministic('mu', a + ba*area + bg*groupsize)
    h = pm.Normal('h', mu=mu, sigma=sigma, observed=weight)
    trace_ag = pm.sample(cores=2)
    
# az.plot_forest(trace_ag, var_names=['a','ba','bg','sigma'], figsize=(6,6), credible_interval=.95)
```

## Model : weight ~ avgfood + groupsize

```python
with pm.Model() as model_fg:
    a = pm.Normal('a', mu=10, sigma=10)
    bg = pm.Normal('bg', mu=0, sigma=1)
    bf = pm.Normal('bf', mu=0, sigma=1)
    sigma = pm.Uniform('sigma', lower=0, upper=10)
    mu = pm.Deterministic('mu', a + bg*groupsize + bf*avgfood)
    h = pm.Normal('h', mu=mu, sigma=sigma, observed=weight)
    trace_fg = pm.sample(cores=2)
    
# az.plot_forest(trace_fg, var_names=['a','ba','bf','sigma'], figsize=(6,6), credible_interval=.95)
```

## Model : weight ~ avgfood + groupsize + area

```python
with pm.Model() as model_afg:
    a = pm.Normal('a', mu=10, sigma=10)
    ba = pm.Normal('ba', mu=0, sigma=1)
    bg = pm.Normal('bg', mu=0, sigma=1)
    bf = pm.Normal('bf', mu=0, sigma=1)
    sigma = pm.Uniform('sigma', lower=0, upper=10)
    mu = pm.Deterministic('mu', a + ba*area + bf*avgfood + bg*groupsize)
    h = pm.Normal('h', mu=mu, sigma=sigma, observed=weight)
    trace_afg = pm.sample(cores=2)
    
# az.plot_forest(trace_afg, var_names=['a','ba','bf','bg','sigma'], figsize=(6,6), credible_interval=.95)
```

```python
## Plot all the models
model_names = ['ag', 'fg', 'afg']
all_traces = [trace_ag, trace_fg, trace_afg]
all_vars = ['a','ba','bf','bg','sigma']
az.plot_forest(all_traces, model_names=model_names, var_names=all_vars, figsize=(6,6), credible_interval=.95)
```
