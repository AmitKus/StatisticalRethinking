## Problem: 6E3

Dice: 1 (20%) , 2 (25%), 3(25%), 4(30%)

```python
def calculate_information_entropy(probs):
    E = 0
    for prob in probs:
        E -= prob*np.log(prob)
    return E
```

```python
calculate_information_entropy([0.2,0.25,0.25,0.3])
```
